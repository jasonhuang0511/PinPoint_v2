```python
import pandas as pd
import numpy as np

import os

%matplotlib inline
from matplotlib import pyplot as plt
import seaborn as sns
index=13077

weight_df=pd.read_csv("C:\\Users\\jason.huang\\research\\GP_momemntum\\weight_df.csv",index_col=0)
close_df=pd.read_csv("C:\\Users\\jason.huang\\research\\GP_momemntum\\crypto_close.csv",index_col=0)
close_df.index=weight_df.index
weight_df.columns=close_df.columns
pct_df=close_df.div(close_df.shift(1))-1

# origin
pnl=weight_df.shift(2).mul(pct_df)
cumpnl=pnl.mean(axis=1).cumsum()
cumpnl.plot()
```

```python
# long short
pnl=weight_df.shift(2).mul(pct_df)
cumpnl_longshort=pnl.iloc[index:,1].cumsum()


weight_df_long=weight_df.applymap(lambda x: x if x>0 else 0)
weight_df_short=weight_df.applymap(lambda x: x if x<0 else 0)
# long
pnl_long=weight_df_long.shift(2).mul(pct_df)
cumpnl_long=pnl_long.iloc[index:,1].cumsum()
# short 
pnl_short=weight_df_short.shift(2).mul(pct_df)
cumpnl_short=pnl_short.iloc[index:,1].cumsum()

df_all=pd.concat([cumpnl_longshort,cumpnl_long,cumpnl_short])
```

```python
# slow weight moving for BNB

result=pd.DataFrame()
threshold_list=[0.05,0.1,0.15,0.2,0.25,0.3,0.35,0.4,0.45,0.5,0.55,0.6]
    
for m in range(len(threshold_list)):
    threshold=threshold_list[m]
    origin_weight_df=pd.DataFrame(weight_df_long.iloc[:,1].copy())
    slow_weight_df=pd.DataFrame(origin_weight_df.copy())
    origin_pct_df=pd.DataFrame(pct_df.iloc[index:,1].copy())

    for j in range(len(slow_weight_df.columns)):
        for i in range(len(slow_weight_df)):
            if i==0:
                pass
            else:
                before_weight=slow_weight_df.iloc[i-1,j]
                target_weight=origin_weight_df.iloc[i,j]
                if target_weight>before_weight:
                    if target_weight>before_weight+threshold:
                        slow_weight_df.iloc[i,j]=before_weight+threshold
                    else:
                        slow_weight_df.iloc[i,j]=target_weight
                else:
                    if target_weight<before_weight-threshold:
                        slow_weight_df.iloc[i,j]=before_weight-threshold
                    else:
                        slow_weight_df.iloc[i,j]=target_weight

    slow_pnl=slow_weight_df.shift(2).mul(origin_pct_df)
    origin_pnl=origin_weight_df.shift(2).mul(origin_pct_df)
    nv_slow_pnl=slow_pnl.mean(axis=1)
    nv_origin_pnl=origin_pnl.mean(axis=1)
    result.loc[m,'threshold']=threshold
    result.loc[m,'origin_ret']=nv_origin_pnl.mean()
    result.loc[m,'origin_ir']=nv_origin_pnl.mean()/nv_origin_pnl.std()
    result.loc[m,'origin_turnover']=origin_weight_df.diff().abs().mean().mean()
    result.loc[m,'slow_ret']=nv_slow_pnl.mean()
    result.loc[m,'slow_ir']=nv_slow_pnl.mean()/nv_slow_pnl.std()
    result.loc[m,'slow_turnover']=slow_weight_df.diff().abs().mean().mean()
    
    print(f"{threshold} is ok")
```

```python
result['after_fee_slow_return']=result['slow_ret']-result['slow_turnover']*0.03/100
result['after_fee_origin_return']=result['origin_ret']-result['origin_turnover']*0.03/100
result['after_fee_slow_return_annual']=result['after_fee_slow_return']*24*365
result
```
