```python
import os
import pandas as pd

%matplotlib inline
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
file_path='C:\\Users\\jason.huang\\research\\data_mining\\GP\\mining_result\\'
result_path=[x for x in os.listdir(file_path) if 'factor_ic_every_other_row_2022_1014_15' in x]
data_all=pd.DataFrame()
for name1 in result_path:
    file_location=file_path+name1+'\\name_dict.csv'
    data=pd.read_csv(file_location,index_col=0)
    data['path']=file_path+name1
    data_all=pd.concat([data_all,data])
data_all=data_all.reset_index(drop=True)
data_all.shape
```

```python
data_all=data_all[data_all['IR_all_data']>1]
data_all=data_all[data_all['Ret_all_data']>0.02]

data_all=data_all.reset_index(drop=True)
data_all=data_all.sort_values('IR_train_data',ascending=False).reset_index(drop=True)
# data_all=data_all.iloc[:min(len(data_all),50),:]
data_all
```

```python
nv_all=pd.DataFrame()
for i in range(len(data_all)):
    nv_file_location=data_all.loc[i,'path']+'\\all_nv\\'+data_all.loc[i,'Factor']+'.csv'
    nv=pd.read_csv(nv_file_location,index_col=0)
    nv=pd.DataFrame(nv['strategy'])
    nv.columns=[data_all.loc[i,'Formula']]
    nv_all=pd.concat([nv_all,nv],axis=1)
nv_all.index.name=None
nv_all.index=pd.to_datetime(nv_all.index)
nv_all
```

```python
nv_total=pd.DataFrame()
for i in range(len(data_all)):
    nv_file_location=data_all.loc[i,'path']+'\\total_nv\\'+data_all.loc[i,'Factor']+'.csv'
    nv=pd.read_csv(nv_file_location,index_col=0)
    nv=pd.DataFrame(nv['strategy'])
    nv.columns=[data_all.loc[i,'Formula']]
    nv_total=pd.concat([nv_total,nv],axis=1)
nv_total.index.name=None
nv_total.index=pd.to_datetime(nv_total.index)
```

```python
plt.rcParams['figure.figsize'] = [20,10]
plt.plot(nv_total.mean(axis=1))
```

```python
plt.rcParams['figure.figsize'] = [20,10]
plt.plot(nv_total.iloc[:,:5])
```

```python
roll_sharpe=nv_total.diff().rolling(90).mean()/nv_total.diff().rolling(90).std()*15.8
position=roll_sharpe.applymap(lambda x: x if x>0 else 0)
pos_sum=position.sum(axis=1)
for i in range(len(position)):
    for j in range(len(position.columns)):
        if pos_sum.iloc[i]==0:
             position.iloc[i,j]=0
        else:
            position.iloc[i,j]=position.iloc[i,j]/pos_sum.iloc[i]
position.sum(axis=1)
fof_nv=position.shift(3)*(nv_total.diff())
fof_nv=fof_nv.cumsum()
plt.plot(fof_nv.mean(axis=1))
```
