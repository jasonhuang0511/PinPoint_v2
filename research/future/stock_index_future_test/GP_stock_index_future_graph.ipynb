```python
import os
import pandas as pd

%matplotlib inline
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
file_path='C:\\Users\\jason.huang\\research\\data_mining\\GP\\mining_result\\'
result_path=[x for x in os.listdir(file_path) if 'pattern_recognition_every_other_row_2022_1017_16' in x]
data_all=pd.DataFrame()
for name1 in result_path:
    file_location=file_path+name1+'\\name_dict.csv'
    data=pd.read_csv(file_location,index_col=0)
    data['path']=file_path+name1
    data_all=pd.concat([data_all,data])
data_all=data_all.reset_index(drop=True)
data_all.shape
```

```python
data_all.sort_values('IR_all_data')
```

```python
data_all=data_all[data_all['IR_all_data']<-0.5]
data_all=data_all[data_all['Ret_all_data']<100]

data_all=data_all.reset_index(drop=True)
data_all=data_all.drop_duplicates('Formula')

data_all=data_all.sort_values('IR_all_data',ascending=False).reset_index(drop=True)
data_all
```

```python
for i in range(len(data_all)):
    print(data_all.loc[i,"Formula"])
```

```python
nv_all=pd.DataFrame()
for i in range(len(data_all)):
    nv_file_location=data_all.loc[i,'path']+'\\all_nv\\'+data_all.loc[i,'Factor']+'.csv'
    nv=pd.read_csv(nv_file_location,index_col=0)
    nv=pd.DataFrame(nv['strategy'])
    nv.columns=[data_all.loc[i,'Formula']]
    nv_all=pd.concat([nv_all,nv],axis=1)
nv_all.index.name=None
nv_all.index=pd.to_datetime(nv_all.index)
nv_all
```

```python
plt.rcParams['figure.figsize'] = [20,10]
plt.plot(nv_all.mean(axis=1).mul(-1))
```
