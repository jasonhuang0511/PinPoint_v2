```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
threshold=1
threshold_long=threshold
threshold_short=-1*threshold

file_save_path='C:\\Users\\jason.huang\\PycharmProjects\\PinPoint_v2\\research\\future\\Intraday_Interest_Future\\result_pickle\\tick\\'

name_str=f"signal_3_long_{threshold_long}_tick_short_{threshold_short}_tick_oi"

indicator_before_fee=f"{file_save_path}_indicator_df_{name_str}.csv"
nv_before_fee=f"{file_save_path}nv_df_{name_str}.csv"
indicator_after_fee2=f"{file_save_path}indicator_df_{name_str}_after_fee2.csv"
nv_after_fee2=f"{file_save_path}nv_df_{name_str}_after_fee2.csv"
indicator_after_fee1=f"{file_save_path}indicator_df_{name_str}_after_fee1.csv"
nv_after_fee1=f"{file_save_path}nv_df_3_{name_str}_after_fee1.csv"

```

```python
indicator_df=pd.DataFrame()

for threshold in range(40):
    threshold_long=threshold
    threshold_short=-1*threshold
    name_str=f"signal_3_long_{threshold_long}_tick_short_{threshold_short}_tick_oi"

    indicator_before_fee=f"{file_save_path}_indicator_df_{name_str}.csv"
    data=pd.read_csv(indicator_before_fee)
    data=data.iloc[[8,17],[1,4,5,7]]
    data['tickers']=['T.CFE','TF.CFE']
    data['param']=threshold
    data['Fee']='fee0'
    turnover=np.array(data['Turnover'])
    indicator_df=pd.concat([indicator_df,data])
    
    indicator_after_fee2=f"{file_save_path}indicator_df_{name_str}_after_fee2.csv"
    data=pd.read_csv(indicator_after_fee2)
    data=data.iloc[[8,17],[1,4,5,7]]
    data['tickers']=['T.CFE','TF.CFE']
    data['param']=threshold
    data['Fee']='fee2'
    data['Turnover']=turnover
    indicator_df=pd.concat([indicator_df,data])
    
    indicator_after_fee1=f"{file_save_path}indicator_df_{name_str}_after_fee1.csv"
    data=pd.read_csv(indicator_after_fee1)
    data=data.iloc[[8,17],[1,4,5,7]]
    data['tickers']=['T.CFE','TF.CFE']
    data['param']=threshold
    data['Fee']='fee1'
    data['Turnover']=turnover
    indicator_df=pd.concat([indicator_df,data])
indicator_df=indicator_df[['tickers','param','Ret','IR','Calmar','Turnover','Fee']]
indicator_df.index=range(len(indicator_df))
```

```python
fig,ax=plt.subplots(2,3,figsize=(3*4,2*4))
a=indicator_df[indicator_df['Fee']=='fee0']
a=a[a['tickers']=='T.CFE']
ax[0,0].plot(a['param'],a['Turnover'])
a=indicator_df[indicator_df['Fee']=='fee0']
a=a[a['tickers']=='TF.CFE']
ax[1,0].plot(a['param'],a['Turnover'])

a=indicator_df[indicator_df['tickers']=='T.CFE']
ret_df=pd.DataFrame()
a1=a[a['Fee']=='fee0']['Ret']
a1.index=range(len(a1))
a2=a[a['Fee']=='fee1']['Ret']
a2.index=range(len(a2))
a3=a[a['Fee']=='fee2']['Ret']
a3.index=range(len(a3))
ret_df=pd.concat([a1,a2,a3],axis=1)
ret_df.columns=['fee0','fee1','fee2']

ax[0,1].plot(ret_df,label=ret_df.columns)
ax[0,1].legend()

a=indicator_df[indicator_df['tickers']=='T.CFE']
ret_df=pd.DataFrame()
a1=a[a['Fee']=='fee0']['IR']
a1.index=range(len(a1))
a2=a[a['Fee']=='fee1']['IR']
a2.index=range(len(a2))
a3=a[a['Fee']=='fee2']['IR']
a3.index=range(len(a3))
ret_df=pd.concat([a1,a2,a3],axis=1)
ret_df.columns=['fee0','fee1','fee2']

ax[0,2].plot(ret_df,label=ret_df.columns)
ax[0,2].legend()



a=indicator_df[indicator_df['tickers']=='TF.CFE']
ret_df=pd.DataFrame()
a1=a[a['Fee']=='fee0']['Ret']
a1.index=range(len(a1))
a2=a[a['Fee']=='fee1']['Ret']
a2.index=range(len(a2))
a3=a[a['Fee']=='fee2']['Ret']
a3.index=range(len(a3))
ret_df=pd.concat([a1,a2,a3],axis=1)
ret_df.columns=['fee0','fee1','fee2']

ax[1,1].plot(ret_df,label=ret_df.columns)
ax[1,1].legend()

a=indicator_df[indicator_df['tickers']=='TF.CFE']
ret_df=pd.DataFrame()
a1=a[a['Fee']=='fee0']['IR']
a1.index=range(len(a1))
a2=a[a['Fee']=='fee1']['IR']
a2.index=range(len(a2))
a3=a[a['Fee']=='fee2']['IR']
a3.index=range(len(a3))
ret_df=pd.concat([a1,a2,a3],axis=1)
ret_df.columns=['fee0','fee1','fee2']

ax[1,2].plot(ret_df,label=ret_df.columns)
ax[1,2].legend()
```

```python
T_nv_df_all=pd.DataFrame()
TF_nv_df_all=pd.DataFrame()


for threshold in range(40):
    threshold_long=threshold
    threshold_short=-1*threshold
    name_str=f"signal_3_long_{threshold_long}_tick_short_{threshold_short}_tick_oi"

    nv_before_fee=f"{file_save_path}nv_df_{name_str}.csv"
    nv=pd.read_csv(nv_before_fee)
    nv_t=pd.DataFrame(nv.iloc[:,1])
    nv_t.columns=[f"{name_str}_fee0"]
    nv_t=nv_t.add(1)
    nv_tf=pd.DataFrame(nv.iloc[:,2])
    nv_tf.columns=[f"{name_str}_fee0"]
    nv_tf=nv_tf.add(1)
    T_nv_df_all=pd.concat([T_nv_df_all,nv_t],axis=1)
    TF_nv_df_all=pd.concat([TF_nv_df_all,nv_tf],axis=1)
    
    nv_after_fee2=f"{file_save_path}nv_df_{name_str}_after_fee2.csv"
    nv=pd.read_csv(nv_after_fee2)
    nv_t=pd.DataFrame(nv.iloc[:,1])
    nv_t.columns=[f"{name_str}_fee2"]
    nv_tf=pd.DataFrame(nv.iloc[:,2])
    nv_tf.columns=[f"{name_str}_fee2"]
    T_nv_df_all=pd.concat([T_nv_df_all,nv_t],axis=1)
    TF_nv_df_all=pd.concat([TF_nv_df_all,nv_tf],axis=1)
    
    nv_after_fee1=f"{file_save_path}nv_df_3_{name_str}_after_fee1.csv"
    nv=pd.read_csv(nv_after_fee1)
    nv_t=pd.DataFrame(nv.iloc[:,1])
    nv_t.columns=[f"{name_str}_fee1"]
    nv_tf=pd.DataFrame(nv.iloc[:,2])
    nv_tf.columns=[f"{name_str}_fee1"]
    T_nv_df_all=pd.concat([T_nv_df_all,nv_t],axis=1)
    TF_nv_df_all=pd.concat([TF_nv_df_all,nv_tf],axis=1)
T_nv_df_all.index=pd.to_datetime(nv.iloc[:,0])
TF_nv_df_all.index=pd.to_datetime(nv.iloc[:,0])
```

```python
threshold=30
threshold_long=threshold
threshold_short=-1*threshold
name_str=f"signal_3_long_{threshold_long}_tick_short_{threshold_short}_tick_oi"
colums_list=[f"{name_str}_fee0",f"{name_str}_fee1",f"{name_str}_fee2"]

fig,ax=plt.subplots(2,1,figsize=(10,20))
ax[0].plot(T_nv_df_all[colums_list],label=T_nv_df_all[colums_list].columns)
ax[0].legend()
ax[1].plot(TF_nv_df_all[colums_list],label=TF_nv_df_all[colums_list].columns)
ax[1].legend()
```
